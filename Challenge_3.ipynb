```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.cloud import bigquery
client = bigquery.Client()

my_query = """
SELECT * FROM `dsc-team-e-pr-428ee1.mdt_dataset.mdt_data`
"""

df = client.query(my_query).to_dataframe()
df
```

```python
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

df.describe()
```

```python
sns.jointplot(x = 'ServingRSRP', y = 'ServingPCI', data = df)
```

```python
df_X = df[['DistanceKilometers', 'Antenna_Height', 'Height', 'Bandwith', 'RS_dBm', 'CRS_Ports', 'Polygon_Type' , 'Polygon_Heigh',
           'Antenna_Azimuth', 'Antenna_MDT', 'Antenna_EDT', 'Antenna_Gain', 'Antenna_HBW', 'Antenna_VBW']].copy()
df_Y = df[['ServingRSRP']].copy()

# Preprocessing to do
# Encode Polygon_Type column
# Get absolute height values Height + Polygon_Heigh



df_X
```

```python
polygon_types = pd.get_dummies(df_X['Polygon_Type'], drop_first = True)
df_X = df_X.drop(['Polygon_Type'], axis=1)
df_X = pd.concat([df_X, polygon_types], axis=1)
df_X
```

```python
# Principal Component Analysis
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaler.fit(df_X)
scaled_data = scaler.transform(df_X)

from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
#pca = PCA(n_components = 0.85, svd_solver = 'full')
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
#print(scaled_data.shape)
#print(x_pca.shape)

df_comp = pd.DataFrame(pca.components_, columns = df_X.columns)
#print(df_comp.head())
plt.figure(figsize = (12,6))
sns.heatmap(df_comp, cmap = 'plasma', annot = True)
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = df_Y['ServingRSRP'], cmap = 'plasma')

plt.suptitle("RSRP vs Principal Components")
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')

plt.colorbar(label = '')
```
